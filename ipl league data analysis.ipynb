```python
# seasons having maximum number of matches
plt.figure(figsize=(10,5))
sns.countplot(x=dataset["season"])
```

```python
plt.figure(figsize=(10,8))
data=dataset["winner"].value_counts()
sns.barplot(x=data,y=data.index)
```

```python
sns.countplot(dataset["toss_winner"]==dataset["winner"])
```

```python
for year in dataset["season"].unique():
       data= dataset[dataset["season"]==year].groupby(["winner","city"])["id"].count().unstack()
       data.plot(kind="bar",stacked=True,title="for year "+str(year))
       plt.legend(loc="best")
      
```

```python
# matches played in different cities
data=dataset["city"].value_counts()
plt.figure(figsize=(10,10))
sns.barplot(x=data,y=data.index)
```

```python
# matches played in stadium
data=dataset["venue"].value_counts()
plt.figure(figsize=(20,10))
sns.barplot(x=data,y=data.index)
```

```python
# teams of tie matches
dataset.loc[dataset["result"]=="tie"][["team1","team2"]].reset_index()
```

```python
# citiwise match take place
data=dataset["city"].value_counts()
data
```

```python
# name of team having consecutively repeated alphabet eg: Chennai super kings
special_team=set()
for team in dataset["team1"].unique():
    for i in range(0,len(team)-1):
        if(team[i]==team[i+1]):
            special_team.add(team)
print(special_team)
            
        
```

```python
# citywise toss winner
data=dataset.groupby("city")["toss_winner"].count()
print("Maximum time toss winning team is: :",data.loc[data==data.max()])
print("Minimum time toss winning team is: :",data.loc[data==data.min()])
```
